# Objetivo

Segmentar os vendedores por categoria e receita de vendas usando modelos de clusterização.

# Bibliotecas

```python
import pandas as pd
from sklearn import cluster
from sklearn import tree
import seaborn as sns
import matplotlib.pyplot as plt
from yellowbrick.cluster import KElbowVisualizer
```

# Lendo as tabelas

```python
catalogo = "../data/02_processed/"

# Tabela de features
df_raw= pd.read_csv(f"{catalogo}tb_book_sellers.csv")
```

```python
df_raw.info()
print(df_raw.shape)
```

```python
df_raw.head()
```

# Modelo

```python
# Escolhendo uma safra e selecionando as features input para o modelo
df = df_raw[df_raw['data'] == '2018-05-01']
print(f'Shape antes: {df.shape}')

features = [c for c in df.columns if c.startswith('recbruta_') and c.endswith('_prop')]

cols = ['data', 'seller_id'] + features

df = df[cols]

print(f'Shape depois: {df.shape}')
```

```python
df.head()
```

## Agglomerative Clustering

```python
# Ajuste do modelo usando Agglomerative Clustering
model = cluster.AgglomerativeClustering(n_clusters=10)
model.fit(df[features]) # fit: ajuste do modelo a partir das features
```

```python
df_copy = df[features].copy()
df_copy['cluster_id'] = model.labels_
df['cluster_id'] = model.labels_

df[['seller_id', 'cluster_id']]
```

```python
df['cluster_id'].value_counts()
```

```python
# Trasnforma as features de proporção em lista

df_cluster_mean = (
    df_copy
    .groupby('cluster_id')
    .agg({col: 'mean' for col in features})
)

df_cluster_mean
```

```python
# Gráfico para mostrar as categorias com maior proporção média (força) dentro de cada cluster
plt.figure(figsize=(15,6))
sns.heatmap(df_cluster_mean)
plt.title('Média das features por cluster')
plt.show()
```

```python
model = cluster.AgglomerativeClustering()
visualizer = KElbowVisualizer(model, k=(5,20))
visualizer.fit(df[features])
visualizer.show()
```

## Árvore de Devisão

```python
print(f"X: {type(df[features])}")
print(f"y: {type(df['cluster_id'])}")
print(f"Qtd Colunas: {len(df[features].columns)}")
```

```python
# Como não temos clareza de quais categorias são mais importantes dentro de cada cluster, optamos por aplicar uma árvore de decisão
# Isso vai possibilitar entendermos quais features são mais importantes dentro de cada cluster

# Usando árvore de decisão para entender quais features são mais importantes na clusterização
X = df[features]
y = df['cluster_id']

clf = tree.DecisionTreeClassifier(random_state=42)
clf.fit(X, y)
```

```python
clf.feature_importances_
```

```python
feature_importance = pd.Series(clf.feature_importances_, index=features) # Pega a importancia das variáveis e ordena
feature_importance = feature_importance.sort_values(ascending=False)
feature_importance
```

```python
# seleciona as 10 features mais importantes
top_features = feature_importance.index[:10].tolist()

df_cluster_mean = (
    df
    .groupby('cluster_id')
    .agg({col: 'mean' for col in top_features})
)

df_cluster_mean
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(df_cluster_mean)
plt.show()
```

## Loop para cada safra

```python
# Escolhendo as safras e selecionando as features de input para o modelo
date_init = '2018-05-01'
date_end = '2018-09-01'

date_list = sorted(list(df_raw[df_raw['data'].between(date_init, date_end)]['data'].unique()))

df_model_output = []
for date in date_list:
    df = df_raw[df_raw['data'] == date]
    print(f'Shape {date} antes: {df.shape}')

    features = [c for c in df.columns if c.startswith('recbruta_') and c.endswith('_prop')]
    cols = ['data', 'seller_id'] + features
    df = df[cols]
    print(f'Shape {date} depois: {df.shape}')

    # Ajuste usando Agglomerative Clustering
    model = cluster.AgglomerativeClustering(n_clusters=10)
    model.fit(df[features]) # fit: ajuste do modelo a partir das features

    df['cluster_id'] = model.labels_
    print(f'Shape {date} final: {df.shape}')

    df_model_output.append(df)

df_model_output = pd.concat(df_model_output, ignore_index=True)
print(f"Shape final: {df_model_output.shape}")
```

```python
df_model_output.groupby("data")["seller_id"].count()
```

# Salvando os clusters

```python
df_model_output.to_csv("../data/03_model/clustering_output.csv", index=False)
```

# Conclusões

* Temos muitas categorias nesta clusterização, o que dificulta o entendimento de quais categorias são mais representativas em cada cluster, ou seja, torna-se difícil identificar qual a categoria predominante dentro de cada grupo. Por isso, optamos por rodar uma árvore de decisão para identificar quais features possuem maior importância na safra escolhida.
